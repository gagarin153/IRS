### repetition_penalty
–≠—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä –ø—Ä–æ—Å—Ç–æ —à—Ç—Ä–∞—Ñ—É–µ—Ç –ø–æ–≤—Ç–æ—Ä—ã. –ß–µ–º –æ–Ω –≤—ã—à–µ, —Ç–µ–º —Å–∏–ª—å–Ω–µ–µ —à—Ç—Ä–∞—Ñ.

```python
text = "–ú–æ–ª–æ–¥–∞—è –∫–Ω—è–≥–∏–Ω—è –ë–æ–ª–∫–æ–Ω—Å–∫–∞—è –ø—Ä–∏–µ—Ö–∞–ª–∞ —Å —Ä–∞–±–æ—Ç–æ–π –≤ —à–∏—Ç–æ–º –∑–æ–ª–æ—Ç–æ–º –±–∞—Ä—Ö–∞—Ç–Ω–æ–º –º–µ—à–∫–µ"
input_ids = tokenizer.encode(text, return_tensors="pt").to(DEVICE)
model.eval()
with torch.no_grad():
    out = model.generate(input_ids, 
                        do_sample=True,
                        num_beams=5, top_k=50,
                        max_length=300,
                        repetition_penalty=3.5
                        )

generated_text = list(map(tokenizer.decode, out))[0]
print()
print(generated_text)
```

### no_repeat_ngram_size
–ê —ç—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä –Ω–∞–ø—Ä—è–º—É—é –≥–æ–≤–æ—Ä–∏—Ç, —á—Ç–æ –Ω–≥—Ä–∞–º–º—ã —Ç–∞–∫–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –Ω–µ –¥–æ–ª–∂–Ω—ã –ø–æ–≤—Ç–æ—Ä—è—Ç—å—Å—è —Å–æ–≤—Å–µ–º

```python
text = "–ú–æ–ª–æ–¥–∞—è –∫–Ω—è–≥–∏–Ω—è –ë–æ–ª–∫–æ–Ω—Å–∫–∞—è –ø—Ä–∏–µ—Ö–∞–ª–∞ —Å —Ä–∞–±–æ—Ç–æ–π –≤ —à–∏—Ç–æ–º –∑–æ–ª–æ—Ç–æ–º –±–∞—Ä—Ö–∞—Ç–Ω–æ–º –º–µ—à–∫–µ "
input_ids = tokenizer.encode(text, return_tensors="pt").to(DEVICE)
model.eval()
with torch.no_grad():
    out = model.generate(input_ids, 
                        do_sample=True,
                        num_beams=3, top_k=100,
                        max_length=100,
                        repetition_penalty=3.5,
                        no_repeat_ngram_size=2,
                        )

generated_text = list(map(tokenizer.decode, out))[0]
print()
print(generated_text)
```
